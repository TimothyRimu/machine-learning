```python
# import pandas
import pandas as pd

# create a dataframe
df = pd.DataFrame({'Experience': [1, 2, 3, 4, 5],
                     'Salary': [45000, 50000, 60000, 80000, 110000]})

# print the dataframe
df
```

Problem statement: how does experience affect the salary of a person?

```python
## Building a linear regression model

# import the ml library
from sklearn.linear_model import LinearRegression

# create you dependent and independent variables
X = df[['Experience']] # independent variable
y = df['Salary'] # dependent variable

# instantial your model
lr = LinearRegression()

# fit the model
lr.fit(X, y)

# check your model score
lr.score(X, y) # R^2
```

```python
# print model coefficients
lr.coef_ # slope
```

```python
# model intercept
lr.intercept_
```

```python
# 10 years experience
salary = 10 * 16000 + 21000
print('estimated salary for 10 years experience is:', salary)
```

```python
print('estimated salary for 10 years experience is:', lr.predict([[10]]))
```

```python
# Plotting the regression line
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue')
plt.plot(X, lr.predict(X), color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```
